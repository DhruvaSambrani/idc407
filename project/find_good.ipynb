```python
import networkx as nx
import matplotlib.pyplot as plt

graphml_list = [i.strip() for i in open("graphml_list").readlines()]
```

```python
lon = []
lom = []
for graphml in graphml_list:
    g = nx.read_graphml(graphml)
    lon.append(len(g.nodes()))
    lom.append(len(g.edges()))
```

```python
plt.hist(lon, bins=40)
plt.xscale("log")
plt.show()
plt.hist(lom, bins=40)
plt.xscale("log")
plt.show()
```

```python
good = []
for i,v in enumerate(graphml_list):
    if lon[i] > 50 and lom[i] > 100:
        good.append(v)
```

```python
open("good_ones", "w").write("\n".join(good))
```
